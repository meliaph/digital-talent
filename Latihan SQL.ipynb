```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

1. Prepare a list of offices sorted by country, state, city.

```python
print("A LIST OF OFFICES SORTED BY COUNTRY, STATE, AND CITY")
run("""
    SELECT country,state,city
    FROM offices
    ORDER BY country,state,city
    """)
```

5. Report total payments for October 28, 2004.

```python
print("TOTAL PAYMENT FOR OCTOBER 28, 2004")
run("""
    SELECT paymentDate, sum(amount)
    FROM payments
    WHERE paymentDate='2004-10-28' 
    ORDER BY paymentDate DESC
    """)
```

11. What is the average percentage markup of the MSRP on buyPrice?

```python
print("THE PERCENTAGE OR MARKUP ON EACH BUYPRICE:")
run("""
    SELECT productCode,buyPrice,MSRP,((MSRP-buyPrice)/buyPrice)*100 AS MarkupValuePercentage
    FROM products
    """)
```

```python
print("THE AVERAGE PERCENTAGE OF MARKUP ON BUYPRICE FOR ALL PRODUCTS:")
run("""
    SELECT AVG(((MSRP-buyPrice)/buyPrice)*100) AS MarkupPercentage
    FROM products
    """)
```

13. Report the name and city of customers who don't have sales representatives?

14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
print("LIST OF EMPLOYEES AS MANAGER AND VP:")
run("""
    SELECT CONCAT(firstName, ' ', lastName) AS fullName, jobTitle
    FROM employees
    WHERE
        jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'
    """)
```

15. Which orders have a value greater than $5,000?

```python
print("LIST OF ORDERS WITH VALUE GREATER THAN $5,000")
run("""
    SELECT orderNumber,productCode,quantityOrdered,priceEach,orderLineNumber,
    quantityOrdered*priceEach AS OrderValue
    FROM orderdetails
    HAVING OrderValue>5000
    ORDER BY OrderValue DESC
    """)
```
