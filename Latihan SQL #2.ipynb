### One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

1. Report the account representative for each customer.

```python
print("THERE ARE 122 CUSTOMERS COUNTED BY ITS CUSTOMER-NUMBER")
run("""
    SELECT COUNT(customerNumber) AS TotalCustomer
    FROM customers
    """)
```

```python
print("LIST OF CUSTOMER WITH EACH SALES REP:")
run("""
    SELECT 
        customerName,
        salesRepEmployeeNumber,
        CONCAT(e.lastName, ', ', e.firstName) AS SalesRepName
    FROM
        customers c 
            INNER JOIN
        employees e
            ON e.employeeNumber = c.salesRepEmployeeNumber
    ORDER BY salesRepEmployeeNumber
    """)

# There are 22 customers without Sales Representative
```

2. Report total payments for Atelier graphique.

```python
print("TOTAL PAYMENT MADE BY ATELIER GRAPHIQUE REGARDLESS ON ITS STATUS (IN US DOLLARS):")
run("""
    SELECT 
        sum(amount) AS AtelierTotalPayment
    FROM 
        customers AS c
            INNER JOIN
        payments AS p
            ON c.customerNumber = p.customerNumber
    WHERE customerName LIKE '%Atelier%'
    """)
```

4. Report the products that have not been sold.

```python
print("NUMBER OF PRODUCT UNSOLD:")
run("""
    SELECT 
        productCode,
        productName,
        quantityInStock,
        quantityOrdered,
        quantityInStock-quantityOrdered AS RemainingUnsold
    FROM 
        orderdetails o
            LEFT JOIN
        products p 
            USING (productCode)
    ORDER BY quantityOrdered
    """)
```

8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
print("PAYMENTS GREATER THAN $100,000 ARE:")
run("""
    SELECT
        c.customerNumber,
        c.customerName,
        sum(amount) AS TotalPayment
    FROM
        customers c
            INNER JOIN
        payments p
            ON c.customerNumber = p.customerNumber
    GROUP BY c.customerNumber
    HAVING TotalPayment>100000
    ORDER BY TotalPayment DESC
    """)
```

```python
print("VALUE OF ON HOLD ORDERS FOR EACH ORDERNUMBER:")
run("""
    SELECT
        o1.orderNumber,
        sum(o1.quantityOrdered*o1.priceEach) AS OnHoldValue,
        status
    FROM
        orderdetails o1
            INNER JOIN
        orders o2
            ON o1.orderNumber = o2.orderNumber
    WHERE status LIKE '%On Hold%'
    GROUP BY o1.orderNumber
    """)
```

1. List products sold by order date.

```python
run("""
    SELECT
        status,
        COUNT(orderNumber)
    FROM
        orders
    GROUP BY status
    """)

# There are 6 status of order
# NOT SURE which status categorized as SOLD 
# ASSUMPTION: 
# Order with status SHIPPED and RESOLVED are categorized as SOLD
# And those, are being assumed that the payment is COMPLETELY DONE
```

```python
print("LIST OF PRODUCT SOLD BY ORDER DATE:")
print("(ONLY SHIPPED and RESOLVED ORDER BEING LISTED HERE)")
print("(ASSUMED THAT SHIPPED and RESOLVED ORDER HAS FINISHED PAYMENT = SOLD)")
run("""
    SELECT
        orderDate,
        o1.orderNumber,
        o1.productCode,
        p.productName,
        status
    FROM
        orderdetails o1
            INNER JOIN orders o2
                ON o1.orderNumber = o2.orderNumber
            INNER JOIN products p
                ON o1.productCode = p.productCode
    WHERE status = 'Resolved'
    OR status = 'Shipped'
    ORDER BY orderDate
    """)
```

2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
print("ORDER LIST OF 1940 FORD PICKUP TRUCK:")
run("""
    SELECT
        o2.orderDate,
        o1.orderNumber,
        o1.productCode,
        p.productName
    FROM
        orderdetails o1
            INNER JOIN
                orders o2 ON (o1.orderNumber = o2.orderNumber)
            INNER JOIN
                products p ON (o1.productCode = p.productCode)
    HAVING productName = '1940 Ford Pickup Truck'
    ORDER BY orderDate DESC
    """)
```

3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
print("LIST OF ORDER VALUED GREATED THAN $25,000 FOR EACH CUSTOMER")
run("""
    SELECT
        c.customerName,
        o1.orderNumber,
        sum(o2.quantityOrdered*priceEach) AS OrderValue
    FROM
        orders o1
            INNER JOIN
                orderdetails o2 ON (o1.orderNumber = o2.orderNumber)
            INNER JOIN
                customers c ON (o1.customerNumber = c.customerNumber)
    GROUP BY o1.orderNumber
    HAVING OrderValue>25000
    ORDER BY c.customerName, OrderValue DESC
    """)
# Urutan Syntax WHERE - GROUP BY - HAVING - ORDER
```

5. List the names of products sold at less than 80% of the MSRP.

```python
print("LIST OF PRODUCTS SOLD WITH LESS THAN 80% OF MSRP:")
#Simply compare the MSRP with each priceEach of product

run("""
    SELECT
        p.productCode,
        productName,
        MSRP,
        priceEach,
        0.8*MSRP AS '80%MSRP'
    FROM
        products p
            INNER JOIN
        orderdetails o 
            ON (p.productCode = o.productCode)
    HAVING priceEach<0.8*MSRP
    ORDER BY priceEach
    """)

# Urutan Syntax WHERE - GROUP BY - HAVING - ORDER
# MSRP = Manufacturer Suggested Retail Price
```

6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)

```python
print("LIST OF PRODUCTS SOLD WITH 100% MARKUP OR ABOVE:")
#Simply compare the buyPrice with each priceEach of product

run("""
    SELECT
        p.productCode,
        productName,
        buyPrice,
        priceEach,
        2*buyPrice AS '100%MarkupAndMore'
    FROM
        products p
            INNER JOIN
        orderdetails o 
            ON (p.productCode = o.productCode)
    HAVING priceEach>2*buyPrice
    ORDER BY priceEach
    """)

# Urutan Syntax WHERE - GROUP BY - HAVING - ORDER
```

7. List the products ordered on a Monday.

```python
print("LIST OF PRODUCT ORDERED ON MONDAY FOR THE GIVEN PERIOD OF TIME:")
run("""
    SELECT
        o2.orderNumber,
        o1.orderDate,
        DAYNAME(o1.orderDate) AS Day,
        p.productName
    FROM
        orderdetails o2
            INNER JOIN orders o1
                ON o2.orderNumber = o1.orderNumber
            INNER JOIN products p
                ON o2.productCode = p.productCode
    HAVING Day = 'Monday'
    """)

# DAYNAME(Table.orderDate) = 'Monday'
```
